```python
numerical_columns = ['carat', 'depth', 'table', 'price', 'x', 'y', 'z']

num_rows = (len(numerical_columns) // 2) + (len(numerical_columns) % 2)
num_cols = 2

fig = sp.make_subplots(rows=num_rows, cols=num_cols,
                       subplot_titles=numerical_columns,
                       vertical_spacing=0.1)
for i, column in enumerate(numerical_columns):
    row = (i // num_cols) + 1
    col = (i % num_cols) + 1
    trace = go.Histogram(x=df[column], nbinsx=30)
    fig.add_trace(trace, row=row, col=col)

fig.update_layout(height=num_rows * 500, width=num_cols * 400,
                  title_text='Histogram  of Numerical columns',showlegend=False)
fig.update_yaxes(title_text='Frequency')
fig.update_xaxes(title_text='Value')

fig.show()
```

. All numerical columns are not normally distributed

```python
fig = make_subplots(rows=1 ,cols =3,subplot_titles=['cut','color','clarity'])

fig.add_trace(go.Bar(x=df['cut'].value_counts().index,y=df['cut'].value_counts().values),row=1,col=1)
fig.add_trace(go.Bar(x=df['color'].value_counts().index,y=df['color'].value_counts().values),row=1,col=2)
fig.add_trace(go.Bar(x=df['clarity'].value_counts().index,y=df['clarity'].value_counts().values),row=1,col=3)
fig.update_layout({'title':{'text':'univariant on numerical columns',
                                   'x':0.5,'y':0.9,
                                   'font_size':26,
                                   'font_color':'black'}},showlegend=False)
fig.show()
```

.In these categorical columns ideal cut,color G, clarity SI1 are maximum

```python
numerical_columns = ['carat', 'depth', 'table', 'x', 'y', 'z']

num_rows = (len(numerical_columns) // 2) + (len(numerical_columns) % 2)
num_cols = 3

fig = sp.make_subplots(rows=num_rows, cols=num_cols,
                       subplot_titles=numerical_columns,
                       vertical_spacing=0.1)
for i, column in enumerate(numerical_columns):
    row = (i // num_cols) + 1
    col = (i % num_cols) + 1
    trace = go.Scatter(x=df[column],y=df['price'],mode='markers')
    fig.add_trace(trace, row=row, col=col)

fig.update_layout(height=num_rows * 500, width=num_cols * 400,
                  title_text='price vs numerical columns')
fig.update_yaxes(title_text='price')
fig.update_xaxes(title_text=column)


fig.show()
```

. In all these price vs numerical columns price is increasing with carat and variable x

```python
box_fig = px.box(df, x='cut', y='price',
                 title='Box Plot of Price by Cut',
                 labels={'cut': 'Cut Quality', 'price': 'Price'})
box_fig.show()
```

```python

numerical_columns = ['carat', 'depth', 'table', 'price', 'x', 'y', 'z']
numerical_data = df[numerical_columns]
corr_matrix = numerical_data.corr()


heatmap_fig = go.Figure(data=go.Heatmap(
                   z=corr_matrix.values,
                   x=corr_matrix.columns,
                   y=corr_matrix.columns,
                   colorscale='gray',
                   colorbar=dict(title="Correlation"),
                   zmin=-1, zmax=1))

heatmap_fig.update_layout(title="Correlation Heatmap",
                          xaxis=dict(title="Features"),
                          yaxis=dict(title="Features"))
heatmap_fig.show()
```

.There is maximum  correlation with price with carat and variable x

```python
fig = px.scatter_3d(df, x='carat', y='depth', z='price', color='price',
                    title='3D Scatter Plot of Price')
fig.show()
```

.From this plot price is increasing with carat but not with depth
